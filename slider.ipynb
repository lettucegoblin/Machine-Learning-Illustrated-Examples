```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
from IPython.display import clear_output

# Define the function to plot
def plot_sine(offset):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x + offset)
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('sin(x + offset)')
    plt.title('Sine Offset')
    plt.grid(True)
    plt.show()

# Create the slider widget
offset_slider = widgets.FloatSlider(value=0.0, min=-2*np.pi, max=2*np.pi, step=0.1, description='Offset')

# Define the callback function for the slider
def update_plot(change):
    clear_output(wait=True)
    plot_sine(change.new)
    display(offset_slider, display_id='slider')


# Link the slider value to the callback function
offset_slider.observe(update_plot, 'value')

# Display the slider with a persistent display_id
display(offset_slider, display_id='slider')
```
