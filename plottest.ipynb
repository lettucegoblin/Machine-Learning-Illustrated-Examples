```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
# Generate x and y values from -10 to 10
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

# Create a meshgrid from x and y values
X, Y = np.meshgrid(x, y)

# Calculate z values using the hyperbolic sine function
Z = np.sinh(X)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('hyperbolic function: z = sinh(x, y)')
plt.show()
```

```python
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Square ass domain Elliptical Paraboloid')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants for the paraboloid
a = 3
b = 3
radius = 5  # Radius of the disk

# Create polar coordinates r and theta
r = np.linspace(0, radius, 100)
theta = np.linspace(0, 2*np.pi, 100)

# Create a meshgrid for polar coordinates
R, Theta = np.meshgrid(r, theta)

# Convert polar coordinates to Cartesian coordinates for X, Y
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Calculate Z based on the elliptical paraboloid equation
Z = (X**2 / a**2) + (Y**2 / b**2)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')  # Using a color map for better visualization

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Elliptical Paraboloid over a Disk')

# Display the plot
plt.show()
```
